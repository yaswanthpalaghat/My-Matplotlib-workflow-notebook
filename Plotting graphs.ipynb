```python
#customizing graphs
import numpy as np
#x=np.array([1,2,3,4])
x=np.arange(0,5,0.1)
y=x**3
y2=x*2
plt.plot(x,y,color="black",marker="o",linewidth="5",label="x^3")#marker=p,o,+,--,-o
plt.plot(x,y2,color="blue",marker="p",linewidth="2",label="x*2")
plt.axis([0,10,0,200])
plt.grid()
plt.text(2,75,"yash",color="red",size="30")
plt.ylabel("y")
plt.xlabel("x")
plt.title("matplot")
plt.legend()
plt.show()
```

```python
#pie graphs
l=["yash","yashu","sai","yaswanth"]
sizes=[3,4,6,2]
c=["red","blue","green","yellow"]
plt.title("splits among classes")
plt.pie(sizes,colors=c,labels=l,autopct="%.1f%%")
plt.axis("equal")
plt.show()
```

```python
l=["yash","yashu","sai","yaswanth"]
sizes=[3,4,6,2]
c=["red","blue","green","yellow"]
explode=[0.1,0.3,0,0]
plt.title("splits among classes")
plt.pie(sizes,colors=c,explode=explode,labels=l,autopct="%.1f%%",counterclock=False,startangle=100)
plt.axis("equal")
plt.show()
```
